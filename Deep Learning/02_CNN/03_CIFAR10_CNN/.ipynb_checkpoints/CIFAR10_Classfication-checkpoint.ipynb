## 1. 데이터 구성 및 샘플 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
import torchvision
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from tqdm import tqdm
from data_loader import get_loaders
```

```python
train_loader, test_loader = get_loaders(batch_size=64)
```

```python
# CIFAR-10 클래스 이름
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck']

# Normalize 되기 전의 이미지 보여주는 함수
def imshow(img):
    img = img.numpy().transpose((1, 2, 0))
    mean = np.array([0.4914, 0.4822, 0.4465])
    std = np.array([0.2470, 0.2435, 0.2616])
    img = std * img + mean  # De-normalize
    img = np.clip(img, 0, 1)  # [0,1] 범위로 자르기
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# 배치에서 샘플 가져오기
dataiter = iter(train_loader)
images, labels = next(dataiter)

# 첫 4개 이미지 출력
for i in range(4):
    imshow(images[i])
    print(f"Label: {classes[labels[i]]}")
```

```python
# 여러 이미지 그리드로 보기
from torchvision.utils import make_grid

def show_grid(img_batch, labels):
    img_batch = img_batch[:16]  # 앞 16개만
    label_batch = labels[:16]
    img_grid = make_grid(img_batch, nrow=4, padding=2)
    imshow(img_grid)
    print('Labels:', [classes[l] for l in label_batch])

show_grid(images, labels)
```

## 2. CNN 모델 설계 및 학습

1) CNN 모델 정의 

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # Convolutional layer
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)    # 입력 채널 3(RGB), 출력 32   
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)   

        # Pooling layer
        self.pool = nn.MaxPool2d(2, 2)

        # Fully connected
        self.fc1 = nn.Linear(64 * 8 * 8, 128)    # CIFAR10 : 32x32 → 8x8 after 2 poolings
        self.fc2 = nn.Linear(128, 10)    # 클래스 수 10개

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))    # Conv1 → ReLU → MaxPool
        x = self.pool(F.relu(self.conv2(x)))    # Conv2 → ReLU → MaxPool
        x = x.view(-1, 64 * 8 * 8)              # Flatten
        x = F.relu(self.fc1(x))                 # Fully connected 1
        x =  self.fc2(x)                        # Fully connected 2 (출력층)
        return x
```

2) CUDA 설정

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

3) 옵티마이저, 손실함수 설정

```python
model = SimpleCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

4) 학습 루프 설정

```python
epochs = 10
# 학습 루프
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # 순전파
        outputs = model(images)
        loss = criterion(outputs, labels)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # 통계
        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    
    # 평가 모드로 전환
    model.eval()
    with torch.no_grad():  # 평가 중에는 기울기 계산 X
        for images, labels in tqdm(test_loader, desc=f"Epoch {epoch+1}/{epochs} - Testing"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = 100 * correct / total

    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {avg_loss:.4f} - Test Accuracy: {acc:.2f}%\n")
```

손실은 줄어들며 정확도는 올라가고 있으므로 학습이 잘 진행되고 있음을 확인할 수 있다.

## 3. 다양한 실험 아이디어 

- 데이터 전처리 다양화  
    `ColorJitter`: 밝기, 채도, 대비 변화  
    `RandomRotation`: 각도 변화

```python
# 새로운 전처리 정의
my_transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),    #  밝기/픽셀 간 대비/채도
    transforms.RandomRotation(15),    # 각도 변화 / 기울어진 이미지에도 강한 성능
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), 
                         (0.2023, 0.1994, 0.2010)),
])

# 테스트셋은 augmentation 없이 normalize만 적용
my_transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), 
                         (0.2023, 0.1994, 0.2010)),
])

# 데이터 로더 불러오기
train_loader, test_loader = get_loaders(batch_size=64, 
                                        transform_train=my_transform_train, 
                                        transform_test=my_transform_test)

epochs = 10
# 학습 루프
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # 순전파
        outputs = model(images)
        loss = criterion(outputs, labels)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # 통계
        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    
    # 평가 모드로 전환
    model.eval()
    with torch.no_grad():  # 평가 중에는 기울기 계산 X
        for images, labels in tqdm(test_loader, desc=f"Epoch {epoch+1}/{epochs} - Testing"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = 100 * correct / total

    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {avg_loss:.4f} - Test Accuracy: {acc:.2f}%\n")
```

1) 기존 모델이 일반화가 잘 되어 있어 정확도가 일정
2) 전처리 강도가 약해서 차이를 만들 수 있음
3) 데이터가 충분히 크고 다양해서 augmentation 효과가 적음

- **CNN 구조 변화**  
    * Conv 레이어 1개 추가

```python
class DeepCNN(nn.Module):
    def __init__(self):
        super(DeepCNN, self).__init__()
        # Convolutional layer
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)    # 입력 채널 3(RGB), 출력 32   
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)

        # Pooling layer
        self.pool = nn.MaxPool2d(2, 2)

        # Fully connected
        self.fc1 = nn.Linear(128 * 4 * 4, 128)    # CIFAR10 : 32x32 → 8x8 after 2 poolings
        self.fc2 = nn.Linear(128, 10)    # 클래스 수 10개

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))    # Conv1 → ReLU → MaxPool
        x = self.pool(F.relu(self.conv2(x)))    # Conv2 → ReLU → MaxPool
        x = self.pool(F.relu(self.conv3(x)))    # Conv3 → ReLU → MaxPool
        x = x.view(-1, 128 * 4 * 4)              # Flatten
        x = F.relu(self.fc1(x))                 # Fully connected 1
        x =  self.fc2(x)                        # Fully connected 2 (출력층)
        return x

model = DeepCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

train_loader, test_loader = get_loaders(batch_size=64)

epochs = 10
# 학습 루프
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # 순전파
        outputs = model(images)
        loss = criterion(outputs, labels)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # 통계
        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    
    # 평가 모드로 전환
    model.eval()
    with torch.no_grad():  # 평가 중에는 기울기 계산 X
        for images, labels in tqdm(test_loader, desc=f"Epoch {epoch+1}/{epochs} - Testing"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = 100 * correct / total

    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {avg_loss:.4f} - Test Accuracy: {acc:.2f}%\n")
```

Conv 레이어를 추가한 후 **73.71%** → **78.47%** 학습 능력이 **올라감** 

- CNN drop out 있는 모델  
    * **Dropout**은 학습 중에 일부 뉴런을 출력 0으로 만들어서 특정 뉴런에 의존하지 않는 방식
    * 사용하는 이유: 과적합 방지, 앙상블과 유사한 효과

```python
class DropoutCNN(nn.Module):
    def __init__(self):
        super(DropoutCNN, self).__init__()
        # Convolutional layer
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)    # 입력 채널 3(RGB), 출력 32   
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)   

        # Pooling layer
        self.pool = nn.MaxPool2d(2, 2)

        # Dropout layer 추가
        nn.Dropout(0.5),    # 학습할 때 50% 확률로 뉴런 꺼버림

        # Fully connected
        self.fc1 = nn.Linear(64 * 8 * 8, 128)    # CIFAR10 : 32x32 → 8x8 after 2 poolings
        self.fc2 = nn.Linear(128, 10)    # 클래스 수 10개

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))    # Conv1 → ReLU → MaxPool
        x = self.pool(F.relu(self.conv2(x)))    # Conv2 → ReLU → MaxPool
        x = x.view(-1, 64 * 8 * 8)              # Flatten
        x = F.relu(self.fc1(x))                 # Fully connected 1
        x =  self.fc2(x)                        # Fully connected 2 (출력층)
        return x

model = DropoutCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

train_loader, test_loader = get_loaders(batch_size=64)

epochs = 10
# 학습 루프
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # 순전파
        outputs = model(images)
        loss = criterion(outputs, labels)

        # 역전파
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # 통계
        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    
    # 평가 모드로 전환
    model.eval()
    with torch.no_grad():  # 평가 중에는 기울기 계산 X
        for images, labels in tqdm(test_loader, desc=f"Epoch {epoch+1}/{epochs} - Testing"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = 100 * correct / total

    print(f"Epoch [{epoch+1}/{epochs}] - Loss: {avg_loss:.4f} - Test Accuracy: {acc:.2f}%\n")
```

정확도 **73.71%** → **74.35%**  
이미 일반화가 잘 되있거나 데이터가 충분해서 **Dropout**의 효과를 크게 못 보고 있다.
