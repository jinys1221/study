```python
# 라이브러리
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## 1. 도형 그리기

```python
# 512 x 512 크기의 검은색 캔버스 생성
canvas = np.zeros((512, 512, 3), dtype=np.uint8)
```

```python
plt.imshow(canvas)
```

- **선**: `cv2.line()`
    - cv2.line(canvas, 시작, 끝, RGB, 두께)

```python
line = cv2.line(canvas, (50, 50), (450, 50), (0, 255, 0), 2)  # 초록색 선
plt.imshow(line)
```

- **사각형**: `cv2.rectangle()`
    - cv2.rectangle(canvas, 왼쪽 상단, 오른쪽 하단, RGB, 두께)

```python
rectangle = cv2.rectangle(canvas, (50, 100), (450, 400), (0, 0, 255), 3)  # 파란색 사각형
plt.imshow(rectangle)
```

- **원**: `cv2.circle()`
    - cv2.circle(canvas, 원의 중심, 반지름, RGB, **두께**)
        - **두께**: -1 원 내부를 채움, 두께 지정 안하면 채워짐

```python
circle = cv2.circle(canvas, (250, 250), 100, (255, 0, 0), 2)  # 빨간색 원
plt.imshow(circle)
```

- **타원**: `cv2.ellipse()`
    - cv2.ellipse(canvas, 타원의 중심, 가로·세로 반지름, 회전 각도, 타원의 시작 각도, 타원의 끝 각도, RGB, 두께)

```python
ellipse = cv2.ellipse(canvas, (250, 250), (100, 50), 20, 0, 270, (255, 255, 0), 2)  # 노란색 타원
plt.imshow(ellipse)
```

## 2. 텍스트 그리기

- **텍스트**: `cv2.putText()`
    - cv2.putTexxt(canvas, 텍스트, 위치, 폰트 종류, 폰트 크기, RGB, 두께)

```python
Text = cv2.putText(canvas, 'Hello, OpenCV!', (100, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
plt.imshow(Text)
```

## 3. 마스킹
마스크는 특정 부분 추출하거나 변경할 때 사용

```python
# 이미지 불러오기
img = cv2.imread("D:/study/OpenCV study/image/baby.jpg")

# 이미지 크기 얻기
height, width = img.shape[:2]

# 마스크 생성 (이미지 크기에 맞게)
mask = np.zeros((height, width), dtype=np.uint8)

# 타원 마스크 생성
mask = cv2.ellipse(mask, (width // 2, height // 2), (200, 100), 340, 0, 360, (255), -1)

# 마스크 적용
masked_image = cv2.bitwise_and(img, img, mask=mask)

# 출력
plt.imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
