```python
# 라이브러리
import cv2
import matplotlib.pyplot as plt
```

## 1. 이미지 불러오기
- `cv2.imread()`: 저장된 경로의 이미지 불러오는 함수, 이미지는 무조건 **BGR**로 읽어온다.

```python
img = cv2.imread("D:/study/OpenCV study/image/dog.jpg")
```

## 2. 이미지 저장
- `cv2.imwrite()`: 입력한 경로에 이미지를 저장하는 함수

```python
import numpy as np
random_img = np.random.randint(0, 256, (256, 256, 3))
cv2.imwrite("D:/study/OpenCV study/01_image_basics/save image/random_image.png", random_img)
```

## 2. 이미지 색상 변환
- `cv2.cvtColor()`: 이미지의 색상 채널 순서 변환하는 함수     
    * `rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`: **BGR → RGB**
    * `gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)`: **BGR → Grayscale**(흑백 이미지)
    * `hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)`: BGR → HSV(색조 기반 탐지에 유용)
    * `lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)`: RGB → LAB (조명 변화에 강한 색공간)

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV는 BGR → RGB 변환 필요

# 시각화
plt.imshow(rgb, cmap="gray")
plt.title("rgb Dog Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 시각화
plt.imshow(gray, cmap="gray")
plt.title("gray Dog Image")
plt.axis("off")
plt.show()
```

## 3. 이미지 크기 변환 & 자르기
1) `cv2.resize()`: 이미지 크기 변경

```python
img = cv2.imread("D:/study/OpenCV study/image/dog.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

resized = cv2.resize(img, (300, 100))

# 시각화
plt.imshow(resized)
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

2) 배열 슬라이싱: `img[y1:y2, x1:x2]`: 이미지 자르기 

```python
cropped1 = img[:500, :200]
cropped2 = img[200:400, 150:350]

# 1행 3열 subplot 구성
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 시각화
axes[0].imshow(img)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(cropped1)
axes[1].set_title("Cropped 1")
axes[1].axis("off")

axes[2].imshow(cropped2)
axes[2].set_title("Cropped 2")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
