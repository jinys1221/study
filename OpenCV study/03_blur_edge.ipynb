```python
# 라이브러리
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## 1. 이미지 불러오기

```python
img = cv2.imread("D:/study/OpenCV study/image/baby.jpg")
```

## 2. 이미지 색상 변환
OpenCV은 **BGR**로 받아들이기 때문에 **RGB**로 변환

```python
Original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 3. 이미지 블러 처리
`cv2.GaussianBlur()`: 이미지의 각 픽셀을 주변 픽셀들과 가우시안 분포에 따른 가중 평균으로 대체
- cv2.GaussianBlur(img, (3, 3), 0)  
    - (3, 3)은 **커널의 크기, 커질수록 강한 블러 효과**
    - **0**은 **표준편차**를 의미한다.
`cv2.medianBlur()`: 각 픽셀을 주변 픽셀들의 중앙값으로 대체
- cv2.medianBlur(img, kernel size)
## 요약
- **가우시안 블러**는 부드러운 흐림 효과, **메디안 블러**는 노이즈 제거 및 경계 보존  

```python
GB33 = cv2.GaussianBlur(Original, (3,3), 0)
GB77 = cv2.GaussianBlur(Original, (7,7), 0)
GB1515 = cv2.GaussianBlur(Original, (15, 15), 0)

MedianBlur = cv2.medianBlur(Original, 11)
```

## 4. 엣지 검출
`cv2.Canny()`  
- cv2.Canny(img, threshold1, threshold2) : **threshold1·2**는 **하위·싱위 임계값**을 의미
    - **하위 임계값**
        - 하위 임계값 이하의 값은 엣지로 인정 x
        - 너무 낮으면 noise가 검출
    - **상위 임계값**
        - 상위 임계값 이상의 값은 엣지로 인정 o

```python
Edge = cv2.Canny(GB1515, 30, 120)
```

## 5. 시각화 비교

```python
plt.figure(figsize=(10, 8))

# 원본 이미지
plt.subplot(2, 3, 1)
plt.imshow(Original)
plt.title('Original')
plt.axis('off')

# 가우시안 블러 3x3
plt.subplot(2, 3, 2)
plt.imshow(GB33)
plt.title('Gaussian Blur 3x3')
plt.axis('off')

# 가우시안 블러 7x7
plt.subplot(2, 3, 3)
plt.imshow(GB77)
plt.title('Gaussian Blur 7x7')
plt.axis('off')

# 가우시안 블러 15x15
plt.subplot(2, 3, 4)
plt.imshow(GB1515)
plt.title('Gaussian Blur 15x15')
plt.axis('off')

# 메디안 블러
plt.subplot(2, 3, 5)
plt.imshow(MedianBlur)
plt.title('MedianBlur')
plt.axis('off')

# 엣지 검출
plt.subplot(2, 3, 6)
plt.imshow(Edge)
plt.title('Edge')
plt.axis('off')

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

## **해석**
- **가우시안 블러**가 강해질수록 이미지의 선명함은 사라지고, 점점 흐려진다.
- **메디안 블러**는 노이즈를 제거하면서도 이미지의 주요 형태와 윤곽을 비교적 잘 유지한다.
- **Canny 엣지 검출**은 가우시안 블러 처리된 이미지에서 강한 경계선을 잘 찾아낸다.
